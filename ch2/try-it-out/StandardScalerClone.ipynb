```python
# Try it out 6
import numpy as np
import pandas as pd
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.utils.validation import check_is_fitted, check_array


class StandardScalerClone(BaseEstimator, TransformerMixin):
    # init
    def __init__(self, with_mean=True):  # no *args or **kwargs!
        self.with_mean = with_mean

    def fit(self, X, y=None):
        if hasattr(X, "columns"):  # it's DataFrame?
            self.feature_names_in_ = X.columns.to_numpy()
        X = check_array(X)
        self.mean_ = X.mean(axis=0)
        self.scale_ = X.std(axis=0)
        self.n_features_in_ = X.shape[1]
        return self

    def transform(self, X):
        check_is_fitted(self)
        X = check_array(X)
        assert self.n_features_in_ == X.shape[1]
        if self.with_mean:
            X = X - self.mean_
        return X / self.scale_

    def inverse_transform(self, X):
        check_is_fitted(self)
        X = X * self.scale_
        if self.with_mean:
            X = X + self.mean_
        return X

    def get_feature_names_out(self):
        return self.feature_names_in_


X = pd.DataFrame({"age": [1.0, 2.0, 3.0], "fare": [100.0, 200.0, 300.0]})
scaler = StandardScalerClone()
X_scaled = scaler.fit_transform(X)
print("scaled:", X_scaled)

X_inversed = scaler.inverse_transform(X_scaled)
print("inversed:", X_inversed)

print(scaler.get_feature_names_out())
```
