```python
import numpy as np
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# X — входные данные (features)
# Каждая строка = один объект
# Каждый элемент внутри = признак объекта
X = np.array([
    [-1, -1],
    [-2, -1],
    [ 1,  1],
    [ 2,  1]
])

# Y — правильные ответы (labels / target)
# Модель должна научиться:
# отрицательные значения -> класс 1
# положительные -> класс 2
Y = np.array([1, 1, 2, 2])

# Pipeline:
# 1. StandardScaler нормализует данные
#    SGD лучше работает когда числа одного масштаба
#
# 2. SGDClassifier обучает модель
#    SGD (stochastic gradient descent)
#    = модель постепенно подкручивает веса,
#      уменьшая ошибку предсказаний
clf = make_pipeline(
    StandardScaler(),
    SGDClassifier(max_iter=1000, tol=1e-3)
)

# fit(X, Y)
# Обучение:
# модель ищет закономерность между X и Y
#
# Упрощенно:
# "какие признаки обычно соответствуют какому классу?"
clf.fit(X, Y)

# predict()
# модель получает новые данные
# и пытается определить класс
print(clf.predict([
    [-0.8, -1],  # похоже на отрицательные -> класс 1
    [30, 1],     # сильно положительное -> класс 2
    [-1, 1],     # смешанный случай
    [-2, -1]     # похоже на класс 1
]))
```
